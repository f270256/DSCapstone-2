Since we have so much missing data from 'us pledged' field and most existing data in that column matches the data in usd_pledge_real, I'll drop the 'us pledged' column.

```python
#Drop 'usd pledged' column
df.drop('usd pledged', axis=1, inplace=True)
```

It looks like deadline and launched should be datetime data types. I'll convert them.

```python
# Convert 'deadline' and 'launched' to datetime objects
df['deadline'] = pd.to_datetime(df['deadline'])
df['launched'] = pd.to_datetime(df['launched'])
```

```python
df.dtypes
```

Let's examine the state column

```python
# Checking unique values and their counts in the 'state' column

df['state'].value_counts()
```

I'm interested in looking at successful vs failed projects so I'll remove the other states

```python
df = df[df['state'].isin(['successful', 'failed', 'canceled', 'undefined', 'suspended'])]
```

```python
# Confirming that the other columns have been removed.

df['state'].value_counts()
```

Let's take a look at the country column next.

```python
# Get the unique values and their counts in the 'country' column
df['country'].value_counts()
```

It looks like there are some countries marked as "N,O'". I'll remove such rows since it's a small percentage of the totals.

```python
df = df[df['country'] != 'N,0"']
```

```python
df['country'].value_counts()
```

```python
df.head()
```

It would be helpful to get the duration of each project's campaign.

```python
# Calculate the duration of the campaign in days
df['duration'] = (df['deadline'] - df['launched']).dt.days
```
